````python
import random
#while迴圈 不知道執行次數+
Max =100
Min = 0
#ans = 56
ans= random.randint(1,99)  # 
count =0

while True:
    print('請輸入',Min,'< <',Max,'範圍的數字',end='')
    guess = int(input("請猜數字"))
    #gues在min mas範圍內
    if guess > Min and guess < Max:
    #原來猜數字的程式
        if guess == ans:
            print("答對了")
            break
        elif guess > ans:
            Max = guess
            print('大於，請繼續猜，您猜了',count,'次')
        else:
            Min= guess
            print('小於，請繼續猜，您猜了',count,'次')
        
        
    else:
    #沒在範圍內則繼續輸入數字
        print("out of range")
    count +=1
````

````python
def convertInputString():
    rawInput= input("\nPlease enter a word, phrase, or a sentence \nto check if it is a palindrome: ")
    rawString=rawInput.lower()
    rawList = list(rawString) #轉成list
    return rawList  #處理過後的, return 後要有變數去儲存

def stripAnalphabetics(dirtyList): #整理成乾淨的List
    analphabeticList=[" ", "-", ".", ",", ":", ";", "!", "?", "'", "\""]
    for character in analphabeticList:
        if character in dirtyList:
            dirtyList.remove(character)
            return stripAnalphabetics(dirtyList)
    return dirtyList
def runPalindromeCheck(straightList):
    reversedList = straightList[::-1] #reverse
    if reversedList == straightList:
        return "the text you have entered is a palindrome!"
    else:
        return "the text you have entered isn't a palindrome."
    

    

originalList = convertInputString()
originalList = stripAnalphabetics(originalList)
palindromeCheck = runPalindromeCheck(originalList)
print(palindromeCheck)
````
